##  Introduction


 dataset: https://www.kaggle.com/datasets/nelgiriyewithana/billionaires-statistics-dataset

###- Project Description
This project focuses on the analysis and classification of a unique dataset containing detailed statistics about the world's billionaires.

The classifications to be performed are threefold:


*   **Wealth Category Classification**: Categorize billionaires into different wealth brackets based on their net worth.
*   **Industry Classification**: Classify billionaires according to the primary industry of their businesses.
*   **Self-Made vs. Inherited Wealt**h: Determine whether the wealth of each billionaire is self-made or inherited.


##  Dataset Selection and Preprocessing

###Importing the libraries

```python
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
```

###Importing the Dataset

```python
dataset = pd.read_csv('Billionaires Statistics Dataset.csv')

dataset.iloc[:, 26] = dataset.iloc[:, 26].str.replace('$', '')
dataset.iloc[:, 26] = dataset.iloc[:, 26].str.replace(',', '')

# Remove rows where any of the data is missing
dataset = dataset.dropna()

# category is the same with industries (exact same)
dataset.drop(columns=dataset.columns[2], inplace=True)

#Rank is not helping in any of the models (also in final worth guess would be a cheat)
dataset.drop(columns=dataset.columns[0], inplace=True)

#personName is not helping in any of the presictions
dataset.drop(columns=dataset.columns[1], inplace=True)

#city is not something I will try to predict or use
dataset.drop(columns=dataset.columns[3], inplace=True)

#city is not something I will try to predict or use
dataset.drop(columns=dataset.columns[3], inplace=True)

#too many different sources to help indicate patterns (difficault with one hot encoding)
dataset.drop(columns=dataset.columns[5], inplace=True)

#droping the birth date since is captured by the birth year, birth month and birth day
dataset.drop(columns=dataset.columns[8], inplace=True)

#Names are not helping in any categorize
dataset.drop(columns=dataset.columns[8], inplace=True)
dataset.drop(columns=dataset.columns[8], inplace=True)

#Title have too many missing values
dataset.drop(columns=dataset.columns[8], inplace=True)

#date of the update of the data unimporntant
dataset.drop(columns=dataset.columns[8], inplace=True)

#the same as the city, I will focus only in the country and not state or residence state
dataset.drop(columns=dataset.columns[8], inplace=True)
dataset.drop(columns=dataset.columns[8], inplace=True)
```

####NN Second Attempt

based on the classification I made we see these:
Ultra Wealth (Category 1): 10 instances (4.20%)
Median Wealth (Category 2): 32 instances (13.45%)
Low Wealth (Category 3): 196 instances (82.35%)

Since this distribution indicates that my classes are indeed imbalanced. The majority of data falls into the "Low Wealth" category.

To address this imbalance, I will consider the following approach:

To create different weight for the different classes

```python
from sklearn.utils.class_weight import compute_class_weight

# Calculate class weights
class_weights = compute_class_weight('balanced',
                                     classes=np.unique(y_Wealth_train),
                                     y=y_Wealth_train)
class_weights = dict(zip(np.unique(y_Wealth_train), class_weights))
```

```python
# Build the neural network
ann2 = tf.keras.models.Sequential([
    tf.keras.layers.Dense(units=6, activation='relu'),
    tf.keras.layers.Dense(units=6, activation='relu'),
    tf.keras.layers.Dense(units=3, activation='sigmoid')  # Assuming 4 output classes
])
```

```python
# Compile the model with the correct loss function for multi-class classification
ann2.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

X_Wealth_train = np.asarray(X_Wealth_train).astype('float32')
X_Wealth_test = np.asarray(X_Wealth_test).astype('float32')

ann2.fit(X_Wealth_train, y_Wealth_train_one_hot, batch_size=32, epochs=100, class_weight=class_weights)
```

```python
ann2.evaluate(X_Wealth_test,y_Wealth_test_one_hot)
```
